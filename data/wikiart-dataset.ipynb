## Preprossesing

```python
artist_class = pd.read_csv("WikiArt/artist_class.txt", delimiter=" ", names=["artist"])
genre_class = pd.read_csv("WikiArt/genre_class.txt", delimiter=" ", names=["genre"])
style_class = pd.read_csv("WikiArt/style_class.txt", delimiter=" ", names=["style"])
artist_train = pd.read_csv("WikiArt/artist_train.csv", delimiter=",", names=["path", "artist"])
genre_train = pd.read_csv("WikiArt/genre_train.csv", delimiter=",", names=["path", "genre"])
style_train = pd.read_csv("WikiArt/style_train.csv", delimiter=",", names=["path", "style"])

artist_train["artist"] = artist_train["artist"].apply(lambda x: artist_class["artist"][x])
genre_train["genre"] = genre_train["genre"].apply(lambda x: genre_class["genre"][x])
style_train["style"] = style_train["style"].apply(lambda x: style_class["style"][x])

print(artist_train.head())
print(genre_train.head())
print(style_train.head())
```

## Data Exploration

```python
print(f"Total: {len(artist_train)}")
artist_train['artist'].value_counts().to_frame()
```

```python
artist_train['artist'].value_counts().plot.barh(xlabel="artist", ylabel="size").set_title("N. of images per artist")
```

```python
print(f"Total: {len(genre_train)}")
genre_train['genre'].value_counts().to_frame()
```

```python
genre_train['genre'].value_counts().plot.barh(xlabel="genre", ylabel="size").set_title("N. of images per genre")
```

```python
print(f"Total: {len(style_train)}")
style_train['style'].value_counts().to_frame()
```

```python
style_train['style'].value_counts().plot.barh(xlabel="style", ylabel="size").set_title("N. of images per style")
```
